# Desafio 1:

## 1 - Realizar a média da metragem para cada um dos bairros.

```python
bairro_metragem = dados[["Bairro","Metragem"]].groupby("Bairro", as_index=False).mean()
bairro_metragem.head(5)
```

```python
# verificando se a media anterior de Aclinação esta correta
tem_imoveis_aclimacao = (dados['Bairro']  == 'Aclimação')
imoveis_aclimacao = dados[tem_imoveis_aclimacao]
imoveis_aclimacao['Metragem'].mean()
```

## 2 - Duas formas de selecionar os dados por bairro

```python
lista_bairros = ["Jardim Guedala", "Vila Mariana", "Ipiranga"]
dados[dados["Bairro"].isin(lista_bairros)].sample(8) # isin() mostra as celulas das colunas lista_birros nao estao vazias
```

```python
dados.loc[dados["Bairro"] == "Alto de Pinheiros"].sample(8) # loc[linha, coluna] localiza uma celula
```

## 3 - Explorar alguns gráficos na documentação e aplicar nas demais colunas do DF (Data Frame), assim como tentar colocar alguma conclusão.

```python
bairros_ordenados_medias = bairro_metragem.sort_values(by="Metragem", ascending=False, ignore_index=True)# sort_values() define ordem dos valores
bairros_ordenados_medias.head(10).plot.bar(x='Bairro', y='Metragem')
```

```python
# Contar quantos imóveis existem nos 10 bairros com maiores médias

bairros_maiores_medias = bairros_ordenados_medias['Bairro'].head(10)
imoveis_bairros_maiores_medias = dados[dados["Bairro"].isin(bairros_maiores_medias)]["Bairro"].value_counts()
display(imoveis_bairros_maiores_medias)
imoveis_bairros_maiores_medias.plot.bar(x='Bairro', y='Metragem')
```

```python
bairros_ordenados_medias.tail(10).plot.bar(x='Bairro', y='Metragem')# mostra as menores medias
```

```python
# Contar quantos imóveis existem nos 10 bairros com menores médias

bairros_menores_medias = bairros_ordenados_medias['Bairro'].tail(10)
imoveis_bairros_menores_medias = dados[dados["Bairro"].isin(bairros_menores_medias)]["Bairro"].value_counts()
display(imoveis_bairros_menores_medias)
imoveis_bairros_menores_medias.plot.bar(x='Bairro', y='Metragem')
```

## 4 - Pegar outras estatísticas dos dados (como média, mediana, mim, max).
